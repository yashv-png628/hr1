```python
hr1_dataset=hr1_dataset.drop('Over18',axis=1)
hr1_dataset=hr1_dataset.drop('EmployeeNumber',axis=1)
hr1_dataset=hr1_dataset.drop('StandardHours',axis=1)
hr1_dataset=hr1_dataset.drop('EmployeeCount',axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
for column in hr1_dataset.columns:
    if hr1_dataset[column].dtype==np.number:
        continue
        hr1_dataset[column]=LabelEncoder().fit_transform(hr1_dataset[column])
    
```

```python
hr1_dataset['Age_Years']=hr1_dataset['Age']
```

```python
hr1_dataset=hr1_dataset.drop('Age',axis=1)
```

```python
hr1_dataset
```

```python
x=hr1_dataset.iloc[:,1:hr1_dataset.shape[1]].values
y=hr1_dataset.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
 
```

```python

from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
forest.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,forest.predict(x_test))

TN=cm[0][0]
TP=cm[1][1]
FN=cm[1][0]
FP=cm[0][1]

print(cm)
print('Model Testing Accuracy= {}' .format((TP+TN)/(TP+TN+FN+FP)))
```
